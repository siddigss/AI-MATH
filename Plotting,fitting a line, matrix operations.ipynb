```python
import numpy as np
import matplotlib.pyplot as plt

# we will plot f(x) = 2x^3 - x om the interval [-3,3]
delta = 0.1
start_point = -5
end_point = 5
x = np.arange(start_point, end_point+delta, delta)
y =   x**3 - 9/2*x**2 +6*x +1
plt.plot(x,y)
```

```python
# The linear regressio eaxmple in the sildes.
points = np.array([[0,0], [1,2], [5,3], [6,5]])
line_info = np.polyfit(points[:,0], points[:,1], 1)
print('line slope = ', line_info[1], ' and its intercept = ', line_info[0])
print('-----------------')
line_limits = np.array([-1,7])
plt.scatter(points[:,0], points[:,1])
plt.plot(line_limits, line_info[0]*line_limits + line_info[1])
```
